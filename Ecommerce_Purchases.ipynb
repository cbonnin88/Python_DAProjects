```python
# How many people with credit cards that expire in 2020 ?
df_purchase['CC Exp Date']
```

```python
def expires() -> int:
  count = 0
  for card in df_purchase['CC Exp Date']:
    if card.split('/')[1]=='20':
      count=count+1
  print(count)
```
